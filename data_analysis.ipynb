```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dontpatronizeme_pcl.tsv', sep="\t", header=None, names=["id + 1", "smth", "keyword", "country", "sentence", "label"])
df
```

```python
df['input_length'] = df['sentence'].astype(str).map(len)
```

```python
df
```

```python
df0 = df.loc[df['label'] == 0]
df1 = df.loc[df['label'] == 1]
df2 = df.loc[df['label'] == 2]
df3 = df.loc[df['label'] == 3]
df4 = df.loc[df['label'] == 4]
```

```python
df0
```

```python
avg_input_length0 = df0["input_length"].mean()
avg_input_length1 = df1["input_length"].mean()
avg_input_length2 = df2["input_length"].mean()
avg_input_length3 = df3["input_length"].mean()
avg_input_length4 = df4["input_length"].mean()
```

```python
print(avg_input_length0, avg_input_length1, avg_input_length2, avg_input_length3, avg_input_length4)
```

```python
list_of_countries = df.country.unique()
list_of_countries.sort()
list_of_countries
```

```python
list_of_keywords = df.keyword.unique()
list_of_keywords.sort()
list_of_keywords
```

```python
df_countries = []
for i in range(len(list_of_countries)):
    df_countries.append(df.loc[df['country'] == list_of_countries[i]])
```

```python
df_countries[0].head()
```

```python
df_keywords = []
for i in range(len(list_of_keywords)):
    df_keywords.append(df.loc[df['keyword'] == list_of_keywords[i]])
```

```python
df_keywords[0].head()
```

```python
classavg_per_country = []
for i in range(len(df_countries)):
    classavg_per_country.append(df_countries[i]['label'].mean())
```

```python
PCLpct_per_country = []
for i in range(len(list_of_countries)):
    PCLpct_per_country.append(float(len(df_countries[i].loc[df_countries[i]['label'] >= 2])) * 100 / len(df_countries[i]))
```

```python
classavg_per_keyword = []
for i in range(len(list_of_keywords)):
    classavg_per_keyword.append(df_keywords[i]['label'].mean())
```

```python
PCLpct_per_keyword = []
for i in range(len(list_of_keywords)):
    PCLpct_per_keyword.append(float(len(df_keywords[i].loc[df_keywords[i]['label'] >= 2])) * 100 / len(df_keywords[i]))
```

```python
dict_country_classavg = {'Country':list_of_countries, 'Class Avg':classavg_per_country}
df_country_classavg = pd.DataFrame(dict_country_classavg)

dict_country_PCLpct = {'Country':list_of_countries, 'PCL %':PCLpct_per_country}
df_country_PCLpct = pd.DataFrame(dict_country_PCLpct)

dict_keyword_classavg = {'Keyword':list_of_keywords, 'Class Avg':classavg_per_keyword}
df_keyword_classavg = pd.DataFrame(dict_keyword_classavg)

dict_keyword_PCLpct = {'Keyword':list_of_keywords, 'PCL %':PCLpct_per_keyword}
df_keyword_PCLpct = pd.DataFrame(dict_keyword_PCLpct)
```

```python
df_country_classavg.sort_values('Class Avg')
```

```python
df_country_PCLpct.sort_values('PCL %')
```

```python
df_keyword_classavg.sort_values('Class Avg')
```

```python
df_keyword_PCLpct.sort_values('PCL %')
```
